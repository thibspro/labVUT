```python
#We implement a half adder in quantum circuit :

#First, we create quantum circuit with 2 qubits and 2 classical bits to store the sum and the carry 
qc = QuantumCircuit(4, 2)
#We apply an X gate to both qubit 
qc.x(0)
qc.x(1)
#We apply a CNOT gate to the first and second qubits
qc.cx(1,2)
qc.cx(0,2)  
#We apply a Toffoli gate
qc.ccx(0,1,3)
qc.measure(2,0)
qc.measure(3,1)
display(qc.draw())     # display a drawing of the circuit
job = sim.run(qc)      # run the experiment
result = job.result()  # get the results
#Interpret the results as a "counts" dictionary
result.get_counts() 
```

```python
#Output explanations : 

#10 represents a sum of 2 in binary, which is the correct result for A = 1 and B = 1 in a half adder. 
#1024 represents the number of shots or repetitions used to run the experiment.
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts())
```

```python
#We check that Hadamard gate changes the state to + state with uncertainty:

qc2 = QuantumCircuit(2, 2)
qc2.h(0)
qc2.measure(0,0)
qc2.measure(1,1)
display(qc2.draw())     # display a drawing of the circuit
job2 = sim.run(qc2)      # run the experiment
result2 = job2.result()  # get the results
# interpret the results as a "counts" dictionary
result2.get_counts()
```

```python
#Output explanations :

#The measurement outcomes were 00 in approximately 514 out of 1024 shots (around 50% probability) 
#The measurement outcomes were 01 in approximately 510 out of 1024 shots (around 50% probability)
```

```python
backend = Aer.get_backend('statevector_simulator')
qc2.x(1)
job3= backend.run(qc2)
result3=job3.result()
outputstate = result3.get_statevector(qc2,decimals=3)
print(outputstate)
```
